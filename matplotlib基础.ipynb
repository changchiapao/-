### 1.Matplotlib图像Figure

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x1 = np.linspace(-5,5,50) #画50个范围在-5到5的等分点
y1 = x1*2
y2 = x1**2 -10
x1
```

```python
plt.plot(x1,y1)
plt.show()
```

```python
plt.figure(figsize = (20,10))
plt.plot(x1,y1)
plt.show()
```

```python
plt.figure(figsize = (20,10))
plt.plot(x1,y1,color = 'blue',linewidth = 6.0,linestyle = 'solid')
plt.plot(x1,y2,color = 'red',linewidth = 3.0,linestyle = 'dashdot')
# linestyle:supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.show()
plt.savefig('./images/x1_y1_y2')
```
